```python
data.rename(columns={'crim': 'CRIM', 'zn': 'ZN', 'indus': 'INDUS', 'chas': 'CHAS', 'nox': 'NOX', 'rm': 'RM', 'age': 'AGE', 'dis': 'DIS','rad':'RAD', 'tax':'TAX','ptratio':'PTRATIO', 'b':'B','lstat':'LSTAT','medv':'MEDV'}, inplace= True)
```

```python
data.head()
```

```python
data['MEDV'] = pd.cut(data['MEDV'], bins=[0,15,30,50], labels=['Low', 'Medium', 'High'])

data.head()
```

```python
tata=data.copy()

tata.head()
```

```python
tata['CHAS'] = tata['CHAS'].replace({1: 'Yes', 0: 'No'})
tata.head()
```

```python
tata.rename(columns={'RM': 'Room'}, inplace=True)
tata.head()
```

**Data Analysis:**

```python
room_medv = tata.groupby('MEDV')['Room'].mean()
room_medv
```

**NumPy Operations:**

```python
age_ary = np.array(data['AGE'])
age_mn = np.mean(age_ary)
age_std = np.std(age_ary)
print("Mean: ", age_mn)
print("Standard Deviation: ", age_std)
```

```python
age_norz = (age_ary - age_mn)/age_std
print("Normalized array: ", age_norz)
```
